```python
import pandas as pd
import os
import json
from pandas.io.json import json_normalize
```

```python
country_counts = {}
```

```python
lender_dir = "loans/"
```

```python
for files in os.listdir(lender_dir):
    inp_file = os.path.join(lender_dir,files)
    print(inp_file)
    with open (inp_file,'r') as inpFile:
        inp_json = json.load(inpFile)
        normalized_json = json_normalize(inp_json,'loans')
        print(normalized_json.columns)
        x = json_normalize(normalized_json['location'])
        print(x.shape)
        print(x.columns)
        m = x.country.value_counts()
        for k,v in m.iteritems():
            if k in country_counts:
                prev_value = country_counts[k]
                new_value = prev_value + v
                country_counts[k] = new_value
            else:
                country_counts[k] = v
        
        
        #for column in normalized_json.columns:
            #print(column,normalized_json[column])
    
    
```
